```python
#Scan the WQ strats, adjust the fees
total_res={}

for ind in ["NASDAQ","DAX","CAC40"]:
    res={}
    for nb in range(1,102):
        try:
            bti=presel.PreselWQ(period,symbol_index=ind, nb=nb)
            bti.run()
            pf=vbt.Portfolio.from_signals(bti.close, bti.entries,bti.exits,short_entries=bti.entries_short,short_exits  =bti.exits_short,freq="1d",#
                                call_seq='auto',cash_sharing=True,fees=0.0001
                                 )
            res[nb]=pf.get_total_return()
        except:
            print("skipped: "+str(nb))
            pass
    total_res[ind]=res
total_res_df=pd.DataFrame(data=total_res)
total_res_df
```
